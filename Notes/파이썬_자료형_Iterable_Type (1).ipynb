```python
#iterable 자료형은 반복이 가능하고, 하나씩 꺼내올 수 있다.
#for 문 과의 호환 좋음
for i in [1, 2, 3]:
  print(i)
```

```python
data_str = '123' #문자열 : 한글자 한글자가 요소 : 순서 ㅇ
data_list = [1, 2, 3] #리스트 : 순서 ㅇ
data_tuple = (1, 2, 3) #튜플 : 순서 ㅇ
data_set = {3, 2, 1} #세트 :
data_dict = {'a':1, 'b':2, 'c' :3} #딕셔너리 key:value 형태 : Key 순서 ㅇ
```

```python
#요소 순서

for i in data_str:
  print(i)

for i in data_list:
  print(i)

for i in data_tuple:
  print(i)

for i in data_set:
  print(i)

for i in data_dict: #Key 만 출력됨
  print(i)


```

```python
#인덱싱

#문자열
data_str[0]
data_str[1]
data_str[2]

#리스트
data_list[0]
data_list[1]
data_list[2]

#튜플
data_tuple[0]
data_tuple[1]
data_tuple[2]

# 셋 : 인덱싱X, 슬라이싱 X
#data_set[0] #error

#딕셔너리 : 슬라이싱X, 키값은 인덱싱 가능
#data_dict[0] #error , put in key, put out value
data_dict['a']
data_dict['b']
data_dict['c']

#슬라이싱
print(data_str[:2], data_list[:2], data_tuple[:2])
```

```python
#요소 할당 =  리스트, 딕셔너리만 가능
#요소 중복 : 문자열, 리스트, 튜플만 가능 (딕셔너리 키 중복 X)

data_list[0] = 100
data_list

data_dict['a'] = 100
data_dict


#요소 추가 = 리스트, 딕셔너리, 집합만 가능

data_str = data_str + '4' #요소 재정의가 아닌 새로운 요소 생성
data_str

data_list.append(4)
data_list

data_set.add(4)
data_set

data_dict['d'] = 400
data_dict


#요소 삭제
data_str = data_str.replace('4', '')
data_str

del data_list[3]
data_list

data_set.discard(4)
data_set

del data_dict['d']
data_dict


#자료형 확인
dir(data_str)
dir(data_list)
dir(data_tuple)
dir(data_set)
dir(data_dict)
```
